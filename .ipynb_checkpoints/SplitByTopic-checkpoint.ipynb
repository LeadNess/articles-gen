```python
from models import MongoStorage, PostgresStorage
```

```python
mdb = MongoStorage.connect(host='localhost')
pg = PostgresStorage.connect(host='172.17.0.2')
```

```python
posts = pg.get_posts(-1)
posts[:1]
```

```python
import pandas as pd

posts_df = pd.DataFrame({
    'id': [post[0] for post in posts],
    'title': [post[1] for post in posts],
    'text': [post[2] for post in posts],
    'date': [post[3] for post in posts],
    'views': [post[4] for post in posts],
    'comments': [post[5] for post in posts],
    'bookmarks': [post[6] for post in posts],
    'rating': [post[7] for post in posts],
    'author': [post[8] for post in posts],
    'habs': [post[9] for post in posts],
    'tags': [post[10] for post in posts],
})
posts_df.head(10)
```

```python
import tqdm

habs_dict = {}

for _, row in tqdm.tqdm(posts_df.iterrows()):
    for hab in row['habs'].split(','):
        if hab in habs_dict:
            habs_dict[hab] += 1
        else:
            habs_dict[hab] = 1
len(habs_dict)
```

```python
habs_dict
```

```python
tags_dict = {}

for _, row in tqdm.tqdm(posts_df.iterrows()):
    for tag in row['tags'].split(','):
        if tag in tags_dict:
            tags_dict[tag] += 1
        else:
            tags_dict[tag] = 1
len(tags_dict)
```

```python
tags_dict
```

```python
import gensim
```

```python
def habs_generator():
    for _, row in posts_df.iterrows():
        yield row['habs'].split(',')
        
def tags_generator():
    for _, row in posts_df.iterrows():
        yield row['tags'].split(',')
```

```python
from gensim.models.phrases import Phraser, Phrases

h_bigram = Phrases(tags_generator())
h_bigram_transformer = Phraser(h_bigram)
```

```python
def tags_generator_bigram():
    for tags in tags_generator():
        yield h_bigram_transformer[ list(tags) ]
```

```python
h_triram = Phrases(habs_generator_bigram())
h_triram_transformer = Phraser(h_triram)
```

```python
def tags_generator_bigram():
    for tags in tags_generator():
        yield h_triram_transformer[ list(tags) ]
```

```python
from gensim.models.word2vec import Word2Vec

h_model = Word2Vec(size=100, window=7, min_count=10, workers=10)
h_model.build_vocab(tags_generator_bigram())
h_model.train(tags_generator_bigram(), total_examples=h_model.corpus_count, epochs=h_model.epochs)
```
