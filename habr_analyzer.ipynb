```python
import pandas as pd

posts_df = pd.DataFrame(data={
    'id': [post[0] for post in en_posts],
    'title': [post[1] for post in en_posts], 
    'text': [post[2] for post in en_posts], 
    'date': [post[3] for post in en_posts],
    'views_count': [post[4] for post in en_posts],
    'bookmarks_count': [post[5] for post in en_posts],
    'comments_count': [post[6] for post in en_posts],
    'rating': [post[7] for post in en_posts],
    'author_nickname': [post[8] for post in en_posts],
    'tags': [post[9] for post in en_posts],
    'habs': [post[10] for post in en_posts]
})
posts_df
```

```python
posts_df = posts_df[(posts_df['title'] != 'Deleted') & (posts_df['text'] != '')]
```

```python
tags = set()
for item in posts_df['tags']:
    for tag in item.split(','):
        tags.add(tag)
len(tags)
```

```python
habs = set()
for item in posts_df['habs']:
    for hab in item.split(','):
        habs.add(hab)
len(habs)
```

```python
tag2posts = {tag: [] for tag in tags}
for index, post in posts_df.iterrows():
    for tag in tags:
        if tag in post['tags']:
            tag2posts[tag].append(post)
tag2posts
```

```python
!mkdir train_en
for tag in tag2posts:
    file_name = tag.replace(' ', '-')
    file_name = file_name.replace('/', '-')
    with open(f'train_en/{file_name}.txt', 'w') as f:
        for post in tag2posts[tag]:
            f.write(post['text'] + '\n<|endoftext|>\n')
        print(f'train_en/{file_name}.txt - completed')
```
